```python
import pandas as pd
import numpy as np
```

```python
# upload the step count data and store it into dataframe
df = pd.read_csv('person1-step-data.csv')
```

```python
# observe the data in the df
df
```

```python
# to check whether exist NULL/missing value in the dataframe
df.isnull().any().any()
```

```python
# observe the data in count column
df["Count"].describe()
```

```python
# group by the data according to the date column
Date_group = df.groupby("Date")
```

```python
# caculate the daily steps count
Date_step_df = Date_group.agg({'Count':'sum'})
Date_step_df
```

```python
# Draw the plot of daily step count from 2014-12-07 ~ 2021-09-22

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

x1 = df["Date"].unique()
y1 = Date_step_df["Count"]

# fig, ax = plt.subplots(1,1)
fig, ax = plt.subplots(figsize=(40,10))
ax.plot(x1,y1)

tick_spacing = 330
ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
plt.ylabel("Step Count")
plt.xlabel("Date")
plt.title("step chart")
plt.grid()
```
